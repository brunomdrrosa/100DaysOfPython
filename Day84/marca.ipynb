# Imports

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

# Abrir imagem

```python
image = Image.open("top.jpg")
image.show()
plt.imshow(image)
```

```python
size = (500, 100)
crop_image = image.copy()
crop_image.thumbnail(size)
```

# Imagem que serÃ¡ a marca d'Ã¡gua

```python
marca = Image.open("logo.png")
marca.show()
plt.imshow(marca)
```

# Marca d'Ã¡gua inserida na imagem escolhida

```python
copied_image = image.copy()
copied_image.paste(marca, (1780, 1050))
plt.imshow(copied_image)
```
